```python
# wikipediaapi
wiki_api = wikipediaapi.Wikipedia(
    language = "en",
    extract_format=wikipediaapi.ExtractFormat.WIKI # for getting data in the string format.

    #extract_format=wikipediaapi.ExtractFormat.HTML for getting data in html format
)
```

```python
corpora = {}
normal = {}
insight = {}
quries = ["electron", "proton", "neutron", "fairy tails", "computer science", "jokes", "new"]

for query in quries:

  small_corpora = {}
  for title in wikipedia.search(query,results=15):
    booloo = False
    try:
      small_corpora[title] = wikipedia.page(title).content
      normal[title] = "done"
    except Exception as error:
      booloo = True
      insight[title] = error
    if booloo :
      small_corpora[title] = wiki_api.page(title).text
  corpora.update(small_corpora)


```

```python
import gensim
```

```python
corpus = {}
for article in corpora:
  corpus[article] = nltk.tokenize.RegexpTokenizer(r"\d*[a-zA-Z]+\d*[a-zA-Z]*").tokenize(corpora[article])
```

```python
corpus.keys()
```

```python
for article in corpus:
  corpus[article] = [words for words in  corpus[article] if len(words)>2]
```

```python
nltk.download('wordnet')
```

```python
for article in corpus:
  corpus[article] = [words.lower() for words in  corpus[article]]
```

```python
for article in corpus:
  corpus[article] = [nltk.stem.wordnet.WordNetLemmatizer().lemmatize(word) for word in  corpus[article]]
```

```python
corpus_values = []
for article in corpus.keys():
  corpus_values.append(corpus[article])
```

```python
len(corpus_values)
```

```python
nltk.download("stopwords")
```

```python
nltk.corpus.stopwords.words("english")
```

```python
new_corpus_values = []
for i in corpus_values:
  sublist=[]
  for word in i:
    if word not in nltk.corpus.stopwords.words("english"):
      sublist.append(word)
  new_corpus_values.append(sublist)
```

```python
bigrames_phrases = gensim.models.phrases.Phrases(new_corpus_values,min_count=5)
```

```python
bigrames_phraser = gensim.models.phrases.Phraser(bigrames_phrases)
```

```python
article_list = list(corpus.keys())
article_list
```

```python
for i in range(len(new_corpus_values)):
  for x in bigrames_phraser[new_corpus_values[i]]:
    if "_" in x:
      new_corpus_values[i].append(x)
      print("{} ============== {}".format(x,article_list[i]))
  
```

```python
corpus_dictionary = gensim.corpora.dictionary.Dictionary(new_corpus_values)
```

```python

corpus_dictionary.filter_extremes(no_below=10,no_above=0.6)
```

```python
print(corpus_dictionary)
```

__LDA model__

```python
# Bag-of-words representation of the documents.
corpus_bag_of_words = [corpus_dictionary.doc2bow(doc) for doc in new_corpus_values]
corpus_bag_of_words
```

```python
lda_model = gensim.models.LdaModel(corpus= corpus_bag_of_words,id2word=corpus_dictionary,
                                   chunksize = 1000,passes = 30,iterations = 600,eval_every = 1,alpha = "auto",eta = "auto",
                                   num_topics = 7)
```

```python
generated_topics = []

for i in range(len(lda_model.top_topics(corpus_bag_of_words))):
  generated_topics.append(lda_model.top_topics(corpus_bag_of_words)[i][0][0][1])

```
